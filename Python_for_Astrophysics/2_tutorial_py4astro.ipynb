### 1D Plotting Example: Salpeter initial mass function (IMF)

The Salpeter power law:

$$\frac{dN}{dM} \propto M^{-2.35}$$

- The Salpeter IMF power law describes the distribution of masses of stars that form in a given region.

- The Salpeter IMF states that the number of stars ($dN$) in a given mass range ($dM$) is proportional to the stellar mass $M$ with index $−2.35$.

- This means there are many more low-mass stars than high-mass stars.

````python
# Salpeter index
alpha = -2.35

# Stellar mass in log10 scale (in solar masses)
M = np.logspace(np.log10(0.1), np.log10(100), 500)

# IMF equation
dN_dM = M**alpha
````

````python
# Create the plot
plt.figure(figsize=(5, 4))

# Plot on a log-log scale
plt.loglog(M, dN_dM, color='darkred', linewidth=2)

plt.xlabel(r'$M/M_{\odot}$')
plt.ylabel(r'$dN/dM$')

plt.show()
````

### 2D Plotting Example:

````python
# Let's define a function using def/return statements

def z_surface(x, y):
    """
    Function that outputs a 2D surface.
    """
    z = np.sin(np.sqrt(x**2 + y**2))

    return z
````

````python
# We need to create a grid, for which we first generate 1D vectors
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
````

````python
# We create a 2D Grid using the vectors above
x_2d, y_2d = np.meshgrid(x, y)
````

````python
# We feed our function:
z_2d = z_surface(x_2d, y_2d)
````

````python
# Plotting with a figure environment
plt.figure(figsize=(8, 6))

ax = plt.axes(projection = "3d")

ax.contour3D(x_2d, y_2d, z_2d, 300)

plt.show()
````
